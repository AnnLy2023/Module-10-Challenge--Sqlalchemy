```python
#Dependencies
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, inspect, text

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Utilize SQLAlchemy create engine function to connect to SQlite database
engine = create_engine("sqlite:///Resources/hawaii.sqlite")
```

```python
# INSPECT
inspector_gadget = inspect(engine)

# get all tables
table_names = inspector_gadget.get_table_names()
for table in table_names:
    print(table)
    
    # get columns for table
    for column in inspector_gadget.get_columns(table):
        print(column)
        
    print()
```

```python
#Create a query that finds the most recent date in the data set (8/23/2017)

query = text("""
                SELECT
                    date,
                    prcp
                FROM
                    measurement
                group by
                    date
                order by
                    date desc
                limit 3000
            """)

pd.read_sql(query, engine)
```

```python
#Create a query that collects only the date and precipitation for the last year of data without passing the date as a variable
#Save the query results to a Pandas DataFrame to create date and precipitation columns

query = text("""
                SELECT
                    date,
                    station,
                    prcp
                FROM
                    measurement
                WHERE
                    date >= '2016-08-23';
               
            """)

df = pd.read_sql(query, engine)
df.head()
```

```python
df.info()
```

```python
#convert Data type for "date"

df["date"] = pd.to_datetime(df["date"])

df.info()
```

```python
#Sort the dataframe by date


df1 = df.dropna().sort_values(by="date", ascending=True).reset_index(drop=True)
df1.head(10)
```

```python

plt.title("Amount of Precipitation in Inches from 8-23-2016 to 8-23-2017")
plt.xlabel("Date")
plt.ylabel("Amount of Precipitation in Inches")
plt.plot(df1.date, df1.prcp, color="orange")
plt.show()
```

# Station Analysis 

```python
#Design a query that correctly finds the number of stations in the dataset (9)
df.station.value_counts()
```

```python
#Design a query that correctly lists the stations and observation counts in descending order and finds the most active station (USC00519281)

query = text("""
                SELECT
                    station,
                    count(*) as num_obs
                FROM
                    measurement
                group by
                    station
                order by
                    num_obs desc;
            """)

pd.read_sql(query, engine)

```

```python
#Design a query that correctly finds the min, max, and avg temp for the most active station (USC00519281)
query = text("""
                SELECT
                    station,
                    min(prcp),
                    avg(prcp),
                    max(prcp)
                FROM
                    measurement
                where
                    station = 'USC00519281';
            """)

pd.read_sql(query, engine)
```

```python
#Design a query to get the previous 12 months of temperature observation (TOBS) data that filters by the station that has the greatest number of observations

query = text("""
                SELECT
                    date,
                    station,
                    tobs
                FROM
                    measurement
                WHERE
                    date >= '2016-08-23'
                    and station = 'USC00519281'
            """)

df3 = pd.read_sql(query, engine)
df3.head()
```

```python

plt.hist(df3.tobs, color="green", bins=12)
plt.title("Temperature Observation from 8-23-2016 to 8-23-2017 for station USC00519281")
plt.ylabel("Frequency")
plt.xlabel("Temperature")
plt.grid()
plt.show()
```
